```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

grubhub_df = pd.read_csv('/content/sentiment_analysis_grubhub.csv')
ubereats_df = pd.read_csv('/content/sentiment_analysis_ubereats.csv')

grubhub_df['Service'] = 'Grubhub'
ubereats_df['Service'] = 'UberEats'
combined_df = pd.concat([grubhub_df, ubereats_df])
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(data=combined_df, x='Sentiment', hue='Service')
plt.title('Sentiment Distribution Comparison')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
grubhub_sentiment_counts = grubhub_df['Sentiment'].value_counts()
plt.pie(grubhub_sentiment_counts, labels=grubhub_sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Grubhub Sentiment Distribution')
plt.subplot(1, 2, 2)
ubereats_sentiment_counts = ubereats_df['Sentiment'].value_counts()
plt.pie(ubereats_sentiment_counts, labels=ubereats_sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('UberEats Sentiment Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=combined_df, x='Polarity', y='Sentiment', hue='Service', style='Service', palette='Set1')
plt.title('Polarity vs Sentiment for Grubhub and UberEats')
plt.xlabel('Polarity')
plt.ylabel('Sentiment')
plt.show()
```
